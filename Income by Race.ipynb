```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# pull in csv
# create variable for race
# create variable for income

race_inc_df = pd.read_csv('../Desktop/Avg Income by Race.csv')
race_inc_df.dropna(inplace = True)
race_inc_df.head()
```

```python
inc_race = race_inc_df
plt.bar(inc_race['Race'],inc_race['Avg Income'])
plt.xlabel("Race", fontsize = 12)
plt.ylabel('Avg Income ($)', fontsize = 12)
plt.xticks(rotation=45)
plt.title("Income by Race")
plt.ylim(0,80000)
plt.show()
```
